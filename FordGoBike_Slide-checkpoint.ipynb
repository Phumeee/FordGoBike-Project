## Investigation Overview

The exploration focused on the effects of trip duration, distance, and bike share for all trip on user type. This presentation begins with an introduction of the individual distribution for the four variables, and then proceeds with the individual and combined effects of the variables on user type. Insights on the strong relationship between trip duration and distance are also displayed. 

The presentation displays the following insights. 91% of the users are subscribers, while 9% are customers. Customers travel for longer hours compared with subscribers, and consequently cover longer distances. This is as a result of the correlation and causal relationship between trip duration and distance, which were moderately positively correlated, and a 1km increase in distance caused a 0.07hr (4.2mins) increase in trip duration. Finally, only subscribers used the bike share for entire trips, and it was ascertained that the majority of the subscribers who used the bike share for the entire trip travelled for smaller durations and distances, with some travelling for longer hours, and covering more distances than others. Some interesting insights outside the determinant variables were also displayed at the end of the presentation.   

## Dataset Overview

The dataset contained 174952 bike share rides with column entries for the user type, trip duration, distance, and bike share for all trip amongst other useful variables like time, gender, and day categories which were briefly displayed at the end of the presentation. A total of 622 datapoints, identified as outliers, were removed during the analysis.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
bike_df = pd.read_csv('Fordgobike_final.csv')
````

````python
# assess the dataset
bike_df.info()
````

## Distribution of User Type

The user type is a categorical variable with two categories namely 'Subscriber', and 'Customer'. 91% of the bike sharers are Subscribers, and the remaining 9% are Customers.


````python
# set color palette
sb.set_palette('Set2') 
````

````python
# create a percentage visualization of the bike sharers by user type
#explode all
explode = (0, 0.1) 
data = pd.DataFrame(bike_df['user_type'].value_counts().reset_index())

#create pie chart
plt.figure(figsize=[6,6])
plt.pie(data=data, x='user_type', labels='index', autopct='%.0f%%', 
        textprops={'fontsize': 16}, explode=explode)

plt.title('% Distribution of Bike Sharers by User type', fontsize=15)
plt.legend(bbox_to_anchor=(1, 1.02), loc='lower left')
plt.show()
````

## Age Distribution

The majority of the bike sharers age between 20 and 40 years, with the peak age at about 31 years.The boxplot and the histogram indicate that the age distribution is right skewed, although unimodal. However, the distribution is logical because, based on the number of trips, younger people should bike more frequently than older people.

````python
# visualize the age distribution
fig, axes = plt.subplots(2, 1, figsize=(16,12))
fig.suptitle('Age Distribution')

plt.subplot(2, 1, 1)
bike_df.boxplot(column=['age'], vert=False);
plt.xlabel('Yrs')

plt.subplot(2, 1, 2)
binsize = 0.5
bins = np.arange(18, bike_df['age'].max()+binsize, binsize)

plt.hist(data = bike_df, x = 'age', bins = bins)
plt.xlabel('Age (Yrs)')
plt.ylabel('No of Members')
plt.show()
````

## Distribution of Trip Duration

The distribution is skewed to the right, although unimodal with a peak point at 0.1hrs (6mins) just like the age distribtuion. The trip duration is concentrated between 0.01hrs (less than 1min) and 0.4hrs (24mins).

````python
# visualize the trip duration distribution
fig, axes = plt.subplots(2, 1, figsize=(16,12))
fig.suptitle('Trip Duration (hr)')

plt.subplot(2, 1, 1)
bike_df.boxplot(column=['duration_hr'], vert=False);
plt.xlabel('Hr')

plt.subplot(2, 1, 2)
binsize = 0.01
bins = np.arange(0.01, bike_df['duration_hr'].max()+binsize, binsize)

plt.hist(data = bike_df, x = 'duration_hr', bins = bins)
plt.xlabel('Trip duration (hr)')
plt.show()
````

## Distribution of Distance

Distance represents the distance from the start to the end stations, and not the entire trip. Its distribution is also right skewed, and seems bimodal with peak points between 1km and 1.6km. 0km represents the distance for trips where the start and end station was the same.

````python
# visualize the distance distribution for the new dataset
fig, axes = plt.subplots(2, 1, figsize=(16,12))
fig.suptitle('Distance (km)')

plt.subplot(2, 1, 1)
bike_df.boxplot(column=['distance'], vert=False);
plt.xlabel('Km')

plt.subplot(2, 1, 2)
binsize = 0.1
bins = np.arange(0., bike_df['distance'].max()+binsize, binsize)

plt.hist(data = bike_df, x = 'distance', bins = bins)
plt.xlabel('Distance (Km)')
plt.show()
````

## Distribution of Bike Use for the Entire Trip

The majority of the bike sharers use the bike for only a part of their trips.

````python
plt.figure(figsize=[8,6])
sb.countplot(data=bike_df, x='bike_share_for_all_trip');

plt.title('Count of Trips based on Bike Use for Entire Trip')
plt.xlabel('Bike Share For Entire Trip')
plt.ylabel('No of Trips')

plt.show()
````

## Differences between User Types based on Trip Duration

Based on the median and quartile, the trip duration for customers is significantly higher than that of subscribers. The range for the trip durations differ significantly for customers (between 0.1hr and 1hr) and subscribers (between 0.1hr and 0.7hr). Furthermore, the probability that the trip duration of subscribers is 0.1hr (roughly 6mins) is very high. On the other hand, customers have a trip duration of 0.25km (roughly 15mins), and the probability is less than than of the subscribers. This is not surprising because subscribers tend to be daily commuters to a defined place of work or study, and some might need to travel longer distances than others.

````python
# create a violin plot for user type vs trip duration
plt.figure(figsize=[8, 6])
sb.violinplot(x=bike_df['user_type'], y=bike_df['duration_hr']);

plt.title('User Type by Trip Duration (Hr)')
plt.xlabel('User Type')
plt.ylabel('Duration (hr)')

plt.show()
````

## Differences between User Types based on Distances

The median of the distance between the start and end stations is slightly higher for customers (1.8km) than for subscribers (1.6km). More customers return the shared bikes to the start station compared to the subscribers. Additionally, the probability density of the subscribers, between 1.5km to 1.7km, is higher compared to that of the customers. It is also interesting that more subscribers return shared bikes to end stations more than 7km away from the start stations, while most customers return bikes to end stations close to 7km away from the start stations.

This insight can also imply the possibility of a correlation between trip duration and distance. This will not be surprising because distance is a part of the trip duration.

````python
plt.figure(figsize=[8, 6])
sb.violinplot(x=bike_df['user_type'], y=bike_df['distance']);

plt.title('User Type by Distance (Km)')
plt.xlabel('User Type')
plt.ylabel('Distance (Km)')

plt.show()
````

## Effect of Distance on Trip Duration

Distance has a significant effect on trip duration. It is interesting to know that about 29% of the total trip duration is determined by the distance between the start and end stations. As earlier stated, this is not surprising because distance is a part of trip duration. From the regression equation, it can be concluded that a 1km increase in distance between start and end station will cause trip duration to increase by 0.06hrs (about 3.6 minutes).

NB: The regression values can change because of the random sampling technique.

````python
from sklearn import linear_model

# select a sample for the scatter plot to avoid overplotting as the population size is over 150000
bike_df_sample = bike_df.sample(n=2000, replace = False)

# identify the data
x = bike_df_sample['distance'].values.reshape(-1,1)
y = bike_df_sample['duration_hr'].values

# create the regression model
ols = linear_model.LinearRegression()
model = ols.fit(x, y)
response = model.predict(x)

# evaluate the r2
r2 = model.score(x, y)

# visualize the regression model

fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, response, color='k', label='Regression model')
ax.scatter(x, y, edgecolor='k', facecolor='green', alpha=0.7, label='Sample data')
ax.set_ylabel('Duration (Hr)', fontsize=14)
ax.set_xlabel('Distance (Km)', fontsize=14)
ax.legend(facecolor='white', fontsize=11)
ax.set_title('$R^2= %.2f$' % r2, fontsize=18)
ax.text(0.55, 0.15, '$y = %.2f x_1 - %.2f $' % (model.coef_[0], abs(model.intercept_)),  fontsize=17, transform=ax.transAxes)

fig.tight_layout()
````

## Differences between User Types based on Bike Share for Entire Trip

It is interesting that only subscribers use the shared bike for an entire trip, while customers use the shared bike for only a part of their trips. The violin plots on user type by trip duration confirmed that subscribers tend to travel shorter distances compared with customers.

````python
plt.figure(figsize=[8, 6])
sb.countplot(data=bike_df,x='bike_share_for_all_trip',hue='user_type')

plt.xlabel('Bike Share for Entire Trip',labelpad=10)
plt.ylabel('Count')
plt.legend(title='User Type')
plt.title('Differences in User Type based on Bike Share for an Entire trip');
````

## Relationship between Trip Duration, Distance and Bike share for all trip for Subscribers

The barplot from the bivariate data exploration established that only subscribers use the bike share for the entire trip. Therefore, the relationship between trip duration, distance and bike share for all trip can be attributed to the subscribers. 

The plot below confirms the assumption that most subscribers travelled to a defined location either for study or work, and some travelled longer distances than others. The concentration of subscribers that used the bike share for the entire trip travelled between 0.2km and about 2km with a trip duration between 0.01 (less than a minute) and 0.2hrs (12mins). Some subscibers also returned the shared bikes to the start station.

````python
g = sb.FacetGrid(data = bike_df_sample, hue = 'bike_share_for_all_trip', size = 5, aspect = 2);
g.map(plt.scatter, 'distance', 'duration_hr');
g.add_legend()

plt.title('Relationship between Trip duration, Distance and Bike Share For all Trip (Subscibers)');
plt.xlabel('Distance (Km)');
plt.ylabel('Duration (hr)');
````

## Additional Insights: Differences between User Types based on Trip Duration and Gender

Although gender had no significant effect on user type when considered as a single variable, it revealed some interesting insights when combined with the trip duration. The females (0.17hrs or 10mins) bike slightly longer than the males (0.14hrs or 8mins), and others (0.16hrs or 9mins) across user types. This insight presents a deviation from current research trends on trip duration between males and females.

````python
plt.figure(figsize=[8,6])
sb.violinplot(data=bike_df,x='gender',y='duration_hr',hue='user_type')

plt.xlabel('Gender', labelpad=15)
plt.ylabel('Duration (hr)')
plt.title('Differences between User types based on Trip duration and Gender')
plt.legend(title='User Type', bbox_to_anchor=(1.3,0.5));
````

## Additional Insights: Differences between User Types based on Trip Duration and Day Category

While day category initially had no effect on user type, an effect was discovered in combination with trip duration. There are significant differences between the trip duration of customers and subscribers on Weekdays and Weekends. While there is no marked difference between the median trip duration for subscribers during the weekdays and weekends, there is a very significant difference between the trip duration for customers on Weekdays and Weekends. The fact that customers tend to ride longer distances compared with subscribers has already been established. Therefore, the key insight here is the fact that more customers use the shared bikes during the weekends, and travel longer distances. This could be for recreation purposes.

NB: Weekends include Saturdays and Sundays, while Weekdays range from Mondays to Fridays.

````python
plt.figure(figsize=[8,6])
sb.violinplot(data=bike_df,x='day_week',y='duration_hr',hue='user_type')
plt.xlabel('Gender', labelpad=15)
plt.ylabel('Duration (Hr)')
plt.title('Differences between User types based on Trip duration and Day Category')
plt.legend(title='User Type', bbox_to_anchor=(1.3,0.5));
````

## Additional Insights: Differences between User Types based on Trip Duration and Time

Finally, Time also had no effect on user type in isolation. In combination with trip duration, it revealed significant differences between the user types. The trip duration for subscribers across the times of the day are more constant at about 0.13hrs (aproximately 8mins). On the other hand, the trip duration for the subscribers fluctuate with a peak point in the afternoons (0.23hrs/approximately 14mins).

NB: morning - 6 to 12; afternoon - 12 to 17; early evening - 17 to 20; late evening - 21 to 23; night - 0 to 6

````python
# order the categories 
cats = [ 'morning', 'afternoon', 'early evening', 'late evening', 'night']
bike_df['time'] = bike_df['time'].astype('category', categories=cats, ordered=True)
````

````python
plt.figure(figsize=[8,6])
sb.violinplot(data=bike_df,x='time',y='duration_hr',hue='user_type')
plt.xlabel('Time', labelpad=15)
plt.ylabel('Duration (Hr)')
plt.title('Differences between User types based on Trip duration and Time')
plt.legend(title='User Type', bbox_to_anchor=(1.3,0.5));
````
